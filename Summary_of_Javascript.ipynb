**Every object has prototype property**

##Object Constructors

```
var john = {
  name: 'John',
  yearOfBirth: 1990,
  job: 'teacher',
  calculateAge = function(){
    console.log();
  });
```
###Better is to write object contructor, and 'john' be instance

```
var Person = function(name, yearOfBirth, job){
  this.name = name;
  this.yearOfBirth = yearOfBirth;
  this.job = job;
  this.calculateAge = function() {
    console.log(2020 - this.yearOfBirth);
  }
};

```
###Create instance (instantiation) of 'john' within blueprint of object constructor

```
var john = new Person ('John', 1990, 'teacher');
```

##Add method to object

```
var jane = new Person('Jane', 1989, 'designer');

```
##Add function to prototype

```
Person.prototype.calculateAge = 
  function() {
    console.log(2020 - this.yearOfBirth);
  };
```

Now call use this as part of object prototype

```
john.calculateAge();

```
